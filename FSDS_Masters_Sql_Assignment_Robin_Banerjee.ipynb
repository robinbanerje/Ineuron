1. Given a table of candidates and their skills, you're tasked with finding the candidates from LinkedIn who are best suited for an open Data Science job. You want to find candidates who are proficient in DS, Python, Tableau, and SQL technology.
Write a query to list the candidates who possess all of the required skills for the job.

Assumption: There are no duplicates in the candidates table.

image.png

```python
'''
## create table: ATS
CREATE TABLE ATS(
id integer,
technology varchar(20)
);
# inserting data into table
INSERT INTO ATS VALUES (1,"DS"), (1,"Tableau"), (1,"SQL"), (2,"R"), (2,"Power BI"), (1,"Python");
'''
# query to list the candidates who possess all of the required technologies
SELECT id FROM ATS WHERE technology = 'DS'
AND id IN (
    SELECT id FROM ATS WHERE technology = 'Tableau'
    AND id IN (
        SELECT id FROM ATS WHERE technology = 'Python'
        AND id IN (
            SELECT id FROM ATS WHERE technology = 'SQL'
        )
    )
);
```

2. Assume you're given two tables containing data about Facebook product pages (product_info table) and their respective likes page (product_info_likes table).

Write a query to return the IDs of the products that have zero likes.

image.png

```python
'''
## create table: product_info
CREATE TABLE product_info(
product_id integer,
product varchar(50)
);
# inserting data into product_info table
INSERT INTO product_info VALUES (1001,"Blog"), (1002,"YouTube"), (1003,"Education");

## create table: likes_info
CREATE TABLE likes_info(
user_id integer,
product_id integer,
liked_date date
);
# inserting data into likes_info table
INSERT INTO likes_info VALUES (1, 1001, '2023-08-19'), (2, 1003, '2023-08-18');
'''
# query to return the IDs of the products that have zero likes
SELECT product_id FROM product_info WHERE product_id NOT IN (SELECT product_id FROM likes_info);
```
